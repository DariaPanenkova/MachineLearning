где:  
gameNo : Последовательный номер матча 

team : Название команды  

oppTeam : Команда - оппозиционер  

matchStage : Этап турнира, на котором игрался матч. (0 - Лига, 1 - Полуфинал, 2 - Финал)  

tossResult :  Результаты броска, чтобы выбрать любую сторону или рейд (0 - Loss, 1 - Win)  

alloutRec : Количество времени, в течение которого команда была полностью уступающей 2 очка   

alloutGiv : Количество времени, в которое команда оппозиции полностью уступала 2 очка  

sTackleRec : Количество раз супер схватки командой, уступающей 2 очка  

sTackleGiv : Количество раз, когда суперпозиция оппозиционной команды давала 2 очка  

touchPntsRec : Количество раз, когда игрок в рейде коснулся соперника, игрок команды приносит 1 очко за каждое прикосновение   

touchPntsGiv : Количество раз, когда игрок оппозиции в рейде коснулся соперника, игрок команды приносит 1 балл за каждое 
прикосновение   

bonusPntsRec : Количество раз, когда игрок в рейде пересек бонусную линию, приносит 1 балл за каждый рейд

bonusPntsGiv :  Количество раз, когда игрок оппозиции в рейде пересек бонусную линию приносит, 1 балл за каждый рейд

raidPntsRec : Количество очков рейда (атаки) по команде, сумма очков касания и бонусных очков

raidPntsGiv : Количество очков рейда (атаки) команды-оппозиции, сумма очков касания и бонусных очков

tacklePntsRec : Количество очков защиты, полученных командой, приносящих 1 очко за обычныый захват и 2 балла за суперзахват 

tacklePntsGiv : Количество очков защиты, полученных командой-оппозиции, приносящих 1 очко за обычныый захват и 2 балла за 
суперзахват 

alloutPntsRec : Колличество All-Out очков, приносящих команде 2 балла (Когда всех игроков команды вытесняют из игры, считается, 
что All-Out навязывается им, а команде оппозиции присуждаются два дополнительных очка.)

alloutPntsGiv : Колличество All-Out очков команды-оппозиции, приносящих 2 балла

extraPntsRec : Количество дополнительных (технических, штрафных) баллов, полученных командой 

extraPntsGiv : Количество дополнительных (технических, штрафных) баллов, полученных командой-оппозицией

totalPntsRec : Количество баллов, полученных командой, сумма очков рейдов, очков схватки, всех очков и дополнительных очков 

totalPntsGiv : Количество баллов, полученных оппозиционной командой, сумма очков рейдов, очков схватки, всех очков и 
дополнительных очков 

touchPntsDiff : Разница touch-очках с оппозиционной командой

bonusPntsDiff : Разница bonus-очках с оппозиционной командой

tacklePntsDiff : Разница в очках защиты с командой оппозиции

alloutPntsDiff : Разница в  all-out очках с командой оппозиции

extraPntsDiff : Разница в лишних баллах

totalPntsDiff : Разница в общих баллах

matchResults : Результат матча (0 - Loss, 1 - Win)

# Постановка задачи 

Определить влияние признаков (таких как результат броска для выбора стороны, наличие бонусных или all-out очков и др.) на достижение победы

Удалим некоторые признаки:'team', 'oppTeam', 'alloutPntsRec', 'alloutPntsGiv', 'touchPntsDiff', 'bonusPntsDiff', 'raidPntsDiff', 'tacklePntsDiff', 'alloutPntsDiff', 'extraPntsDiff', 'totalPntsDiff', т.к. они либо не влияют на достижение победы, либо легко вычисляются с помощью других признаков 

```python
data =  data.drop('team', axis = 1)\
            .drop('oppTeam', axis = 1)\
            .drop('alloutPntsRec', axis = 1)\
            .drop('alloutPntsGiv', axis = 1)\
            .drop('touchPntsDiff',axis = 1)\
            .drop('bonusPntsDiff',axis = 1)\
            .drop('raidPntsDiff', axis = 1)\
            .drop('tacklePntsDiff',axis = 1)\
            .drop('alloutPntsDiff', axis = 1)\
            .drop('extraPntsDiff', axis = 1)\
            .drop('totalPntsDiff',axis = 1)\
            .drop('totalPntsRec',axis = 1)
```

```python
categorical_columns = [c for c in data.columns if data[c].dtype.name == 'object']
numerical_columns   = [c for c in data.columns if data[c].dtype.name != 'object']
print(categorical_columns)
print(numerical_columns)
```

# Выборки

```python
X = data.drop(('matchResult'), axis = 1)
```

```python
Y = data['matchResult']
feature_names = X.columns
feature_names
```

```python
print(X.shape)
print(Y.shape)
N, d = X.shape
```

```python

type(X)
```

```python

type(Y)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

Получили тестовую и обучающую выборки

```python

print(N_train + N_test)
```

# Метод К ближайщих соседей

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)

knn.fit(X_train, Y_train)
```

```python
Y_train_predict = knn.predict(X_train)
Y_test_predict = knn.predict(X_test)

err_train = np.mean(Y_train != Y_train_predict)
err_test  = np.mean(Y_test  != Y_test_predict)

print(err_train, err_test)
```

# Обучение полносвязной (неглубокой) нейронной сети

## Обучаем классификатор

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (100,), solver = 'lbfgs', 
                          activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, Y_train)
```

Проверяем качество классификатора

```python
Y_train_pred = mlp_model.predict(X_train)
Y_test_pred = mlp_model.predict(X_test)
```

Ошибки на обучающей и тестовой выборках:

```python

print(np.mean(Y_train != Y_train_pred), np.mean(Y_test != Y_test_pred))
```
